## Downloading Metadata

```python
# Use libraries to call API, download into JSON format, and convert into CSV.
import json
import pandas as pd
import requests

def save_metadata(results_url):

    params = {"fo": "json", "c": "100", "at": "results,pagination"}
    call = requests.get(results_url, params=params)
    data = call.json()
    results = data['results']
    
    #read json into a file
    with open('metadata.json', 'w') as json_file:
        json.dump(results, json_file)
    
    #convert json to csv
    df = pd.read_json('metadata.json')
    df.to_csv('metadata.csv', index=False)
#TODO: save file to folder called 'metadata'
```
